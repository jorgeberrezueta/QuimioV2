```python
import pandas as pd
import numpy as np
from regression import LinearRegressionGD
```

```python
datos = pd.read_csv('datos.csv')

completos = datos.dropna()

datos.head()
```

```python
completos.iloc[:, 5].values
```

```python
model = LinearRegressionGD()

model.fit(
    completos.iloc[:, 5:].values,
    completos.iloc[:, 3].values
)
```

```python
prediction = model.predict(datos.iloc[:,5:].values)

pd.DataFrame({
    'orig': datos.iloc[:,3].values,
    'pred': prediction,
    'diff': datos.iloc[:,3].values - prediction,
    'perc': (prediction / datos.iloc[:,3].values) * 100
}).dropna()
```
